Injuries and robustness to destruction are a key interest point for the client. **Construct metric for fatal/serious injuries**

Construct a derived column tracking whether an aircraft was destroyed or not.

```python
# Clean Injury Columns
injury_cols = ['Total.Fatal.Injuries', 'Total.Serious.Injuries',
               'Total.Minor.Injuries', 'Total.Uninjured']

print("Null counts in injury columns:")
print(df[injury_cols].isnull().sum())
print("\nDescriptive statistics:")
print(df[injury_cols].describe())
```

```python
# Impute NaN values in injury columns with 0
# Rationale: if no injuries of a type were recorded, it is reasonable to treat as 0
for col in injury_cols:
    df[col] = df[col].fillna(0)

# Estimate total number of people on board
df['total_aboard'] = (df['Total.Fatal.Injuries'] +
                      df['Total.Serious.Injuries'] +
                      df['Total.Minor.Injuries'] +
                      df['Total.Uninjured'])

# Drop rows where no passenger info is available (total_aboard == 0)
df = df[df['total_aboard'] > 0]
print(f"Rows after dropping zero-passenger records: {len(df)}")

# Create the key safety metric: fraction fatally or seriously injured
df['fatal_serious_fraction'] = (
    (df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries']) / df['total_aboard']
)

print("\nfatal_serious_fraction statistics:")
print(df['fatal_serious_fraction'].describe())
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# Drop rows with missing model
df = df.dropna(subset=['Model'])
print(f"Rows after dropping null Model: {len(df)}")

# Check whether model names are unique to each make
print("\nTotal unique models:", df['Model'].nunique())

# Check if same model name appears under multiple makes
model_make_counts = df.groupby('Model')['Make'].nunique()
shared_models = model_make_counts[model_make_counts > 1]
print(f"\nModel names shared across multiple makes: {len(shared_models)}")
print("\nTop 10 shared model names:")
print(shared_models.sort_values(ascending=False).head(10))
```

```python
# Model labels are NOT unique to each make — e.g., '172' could be Cessna or something else.
# Create a 'make_model' identifier combining Make and Model.

df['Model'] = df['Model'].str.strip().str.upper()
df['make_model'] = df['Make'].str.strip() + ' ' + df['Model'].str.strip()

print("Sample make_model values:")
print(df['make_model'].value_counts().head(20))
print(f"\nTotal unique make_model identifiers: {df['make_model'].nunique()}")
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
# Engine.Type
print("Engine.Type values:")
print(df['Engine.Type'].value_counts())
print()

# Standardize casing and strip whitespace
df['Engine.Type'] = df['Engine.Type'].str.strip().str.title()

# Map 'Unknown' / 'None' to NaN
df['Engine.Type'] = df['Engine.Type'].replace({'Unknown': np.nan, 'None': np.nan, 'Nr': np.nan})

print("Engine.Type after cleaning:")
print(df['Engine.Type'].value_counts(dropna=False))
```

```python
# Weather.Condition
print("Weather.Condition values:")
print(df['Weather.Condition'].value_counts(dropna=False))
print()

# Standardize: merge 'Unk' and 'UNK' into NaN
df['Weather.Condition'] = df['Weather.Condition'].str.strip().str.upper()
df['Weather.Condition'] = df['Weather.Condition'].replace({'UNK': np.nan})

print("Weather.Condition after cleaning:")
print(df['Weather.Condition'].value_counts(dropna=False))
```

```python
# Number.of.Engines
print("Number.of.Engines values:")
print(df['Number.of.Engines'].value_counts(dropna=False))

# Keep as float; 0 engines is suspicious but we leave it in and note it.
# Cast to int where not null for clarity (will stay float due to NaN)
print("\nNull count:", df['Number.of.Engines'].isnull().sum())
```

```python
# Purpose.of.flight
print("Purpose.of.flight values:")
print(df['Purpose.of.flight'].value_counts(dropna=False).head(15))
print()

df['Purpose.of.flight'] = df['Purpose.of.flight'].str.strip().str.title()
# Map 'Unknown' to NaN
df['Purpose.of.flight'] = df['Purpose.of.flight'].replace({'Unknown': np.nan})

print("After cleaning (top 10):")
print(df['Purpose.of.flight'].value_counts(dropna=False).head(10))
```

```python
# Broad.phase.of.flight
print("Broad.phase.of.flight values:")
print(df['Broad.phase.of.flight'].value_counts(dropna=False).head(15))
print()

df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].str.strip().str.title()
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].replace({'Unknown': np.nan})

print("After cleaning (top 10):")
print(df['Broad.phase.of.flight'].value_counts(dropna=False).head(10))
```

### Column Removal
- inspect the dataframe and drop any columns that have too many NaNs

```python
# Compute null percentage per column
null_pct = df.isnull().mean() * 100
print("Null percentage per column (sorted):")
print(null_pct.sort_values(ascending=False).round(2))
```

```python
# Drop columns with >= 60% nulls
drop_threshold = 70  # Lowered to 70% to preserve Broad.phase.of.flight (~86% null)
cols_to_drop = null_pct[null_pct >= drop_threshold].index.tolist()
print(f"Columns to drop (>= {drop_threshold}% null): {cols_to_drop}")

# Broad.phase.of.flight has ~86% null but is analytically valuable — keep it
cols_to_drop = [c for c in cols_to_drop if c != 'Broad.phase.of.flight']
df = df.drop(columns=cols_to_drop, errors='ignore')
print(f"\nRemaining columns: {df.columns.tolist()}")
print(f"Final shape: {df.shape}")
```
